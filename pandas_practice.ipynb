```python
df = pd.DataFrame(data)
df
```

```python
x = df['Employee_ID']
y = df['Department']
plt.bar(x,y)
plt.title("Employee_id vs Department")
plt.xlabel("Employee_id")
plt.ylabel("Department")
plt.legend()
plt.show()
```
