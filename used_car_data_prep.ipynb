## **Task 1** Load, Inspect and Clean

```python
import kagglehub
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import *

# Download latest version
path = kagglehub.dataset_download("manishkr1754/cardekho-used-car-data")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("/kaggle/input/cardekho-used-car-data/cardekho_dataset.csv")
df
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.isnull().sum()*100/len(df))
```

```python
df.dropna(subset = ['selling_price'],inplace = True)
```

```python
df['mileage'] = (df['mileage'].astype('str').str.replace('kmpl','').str.strip())
df['mileage'] = pd.to_numeric(df['mileage'],errors = 'coerce')
```

```python
df['engine'] = (df['engine'].astype('str').str.replace('CC','').str.strip())
df['engine'] = pd.to_numeric(df['engine'],errors = 'coerce')
```

```python
df['max_power'] = (df['max_power'].astype('str').str.replace('bhp','').str.strip())
df['max_power'] = pd.to_numeric(df['max_power'],errors = 'coerce')
```

```python
for col in ['mileage','engine','max_power']:
  median_val = df[col].median()
  df[col] = df[col].fillna(median_val)
  print(f"Filled {col} with median : {median_val}")
```

```python
df = df[(df['selling_price'] != 999999999) | (df['selling_price'] >= 10000)]
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
print(df.shape)
```

## **Task 2**  Encode Categorical Features

```python
df['transmission_type'] = df['transmission_type'].map({'Manual': 0, 'Automatic' : 1})
df
```

```python
df = pd.get_dummies(df,columns = ['fuel_type'],drop_first = True)

```

```python
df = pd.get_dummies(df,columns = ['seller_type'],drop_first = True)
```

```python
df.columns.tolist()
```

###### We use drop_first = True in one hot encoding,to reduce multicollinearity, which confuses linear models.

A row of all zeros in the new dummy columns represents the dropped category.

## **Task 3** : Split and Compute Baseline MAE

```python

df = df.drop(columns = ['Unnamed: 0',	'car_name',	'brand',	'model'])
x = df.drop(columns = ['selling_price'])
y = df['selling_price']

```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.80,random_state = 42)
```

```python
my_model = LinearRegression()

my_model.fit(x_train,y_train)

y_pred = my_model.predict(x_test)

mae = mean_absolute_error(y_test,y_pred)
print(f"Baseline MAE: ₹{mae:.2f}")


```
